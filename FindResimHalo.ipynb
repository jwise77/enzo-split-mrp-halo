```python
import yt
import numpy as np
import ytree
```

```python
a = ytree.load('rockstar_halos/trees/tree_0_0_0.dat')
```

```python
a[0]
```

```python
ds = yt.load('DD0125/output_0125')
```

```python
center = ds.arr([0.5]*3, 'code_length')
width = ds.quan(0.1, 'code_length')
reg = ds.region(center, center-0.5*width, center+0.5*width)
```

```python
gfields = [('gas', 'density'), ('gas', 'temperature')]
dfields = [('gas', 'dark_matter_density')]
gweight = ('gas', 'density')
dweight = ('gas', 'dark_matter_density')

for dim in 'xyz':
    proj = yt.ProjectionPlot(ds, dim, gfields, weight_field=gweight, width=width, center=center, data_source=reg)
    proj.show()
    dmproj = yt.ProjectionPlot(ds, dim, dfields, weight_field=dweight, width=width, center=center, data_source=reg)
    dmproj.show()
```

```python
min_mass = a.quan(1e6, 'Msun')
fil = a['mass'] >= min_mass
fil_ids = np.where(fil)[0]
```

```python
halo_fields = ['mass', 'position', 'virial_radius', 'id']
for i in fil_ids:
    print(f"Halo {a['id'][i]:5d}: M_h = {a['mass'][i]:.3g}, pos = {a['position'][i]}, rvir = {a['virial_radius'][i]*a['scale'][i]:.3f}")
```

```python
focus_id = 26813
focus_i = np.where(a['id'] == focus_id)[0][0]
pos = ds.arr(a['position'][focus_i].to('Mpc'), 'Mpccm')
rvir = ds.quan(a['virial_radius'][focus_i].to('kpc'), 'kpccm')
mass = ds.quan(a['mass'][focus_i].to('Msun'), 'Msun')
```

```python
hwidth = ds.quan(4*rvir)
hreg = ds.region(pos, pos-0.55*hwidth, pos+0.55*hwidth)
```
