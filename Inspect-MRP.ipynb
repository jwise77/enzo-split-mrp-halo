```python
import yt
yt.enable_plugins()
```

```python
num = 59
```

# New run with MRP only in halo of interest

```python
ds = yt.load(f"DD{num:04d}/output_{num:04d}")
ds.add_particle_filter('mrp')
```

```python
rhomax, center = ds.find_max(('gas', 'density'))
width = ds.quan(3, 'kpc')
```

```python
for dim in 'xyz':
    p = yt.ProjectionPlot(ds, dim, ('deposit', 'mrp_density'), width=width, center=center)
    p.show()
    p.save()
```

```python
for dim in 'xyz':
    p = yt.ProjectionPlot(ds, dim, ('gas', 'dark_matter_density'), weight_field=('gas', 'dark_matter_density'), width=width, center=center)
    p.show()
    p.save()
```

```python
for dim in 'xyz':
    p = yt.ProjectionPlot(ds, dim, ('gas', 'density'), weight_field=('gas', 'density'), width=width, center=center)
    p.show()
    p.save()
```

```python
for dim in 'xyz':
    p = yt.ProjectionPlot(ds, dim, ('index', 'grid_level'), method='max', width=width, center=center)
    p.set_cmap('all', 'tab10')
    p.set_log('all', False)
    p.show()
    p.save()
```

# Checking for massive particle pollution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp = ds.sphere(center, ds.quan(500, 'pc'))
```

```python
pmass = sp['nbody', 'particle_mass']
```

```python
pmass.to('Msun').mean()
```

```python
np.median(pmass.to('Msun'))
```

```python
plt.hist(np.log10(pmass.to('Msun')))
```

```python
p = yt.ProjectionPlot(ds, 'x', ('deposit', 'all_density'), weight_field=('deposit', 'all_density'), width=2*sp.radius, center=center, data_source=sp)
p.show()
p.save()
p = yt.ProjectionPlot(ds, 'x', ('gas', 'density'), weight_field=('gas', 'density'), width=2*sp.radius, center=center, data_source=sp)
p.show()
p.save()
```
