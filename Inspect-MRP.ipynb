```python
import yt
yt.enable_plugins()
```

```python
num = 54
```

# New run with MRP only in halo of interest

```python
ds = yt.load(f"DD{num:04d}/output_{num:04d}")
ds.add_particle_filter('mrp')
```

```python
center = ds.arr([0.5]*3, 'unitary')
width = ds.quan(0.25, 'unitary')
```

```python
#rhomax, center = ds.find_max(('gas', 'density'))
#width = ds.quan(3, 'kpc')
```

```python
p = yt.ProjectionPlot(ds, 'x', ('deposit', 'mrp_density'), width=width, center=center)
p.show()
p.save()
```

```python
p = yt.ProjectionPlot(ds, 'x', ('deposit', 'all_density'), width=width, center=center)
p.show()
p.save()
```

```python
p = yt.ProjectionPlot(ds, 'x', ('gas', 'density'), weight_field=('gas', 'density'), width=width, center=center)
p.show()
p.save()
```

```python
p = yt.ProjectionPlot(ds, 'x', ('index', 'grid_level'), method='max', width=width, center=center)
p.set_cmap('all', 'tab10')
p.set_log('all', False)
p.show()
```
